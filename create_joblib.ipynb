```python
import os
import time
import joblib
import librosa
import numpy as np
```

```python
#creates 2 joblib files seperately for data and labels
def features_creator(path, save_dir) -> str:
    lst = []
    start_time = time.time()

    for subdir, dirs, files in os.walk(path):
        for file in files:
            try:
                # Load librosa array, obtain mfcss, store the file and the mcss information in a new array
                X, sample_rate = librosa.load(os.path.join(subdir, file),
                                                res_type='kaiser_fast')
                mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate,
                                                        n_mfcc=40).T, axis=0)
                # The instruction below converts the labels (from 1 to 5) to a series from 0 to 4
                # This is because our predictor needs to start from 0 otherwise it will try to predict also 0.
                file = int(file[7:8]) - 1
                arr = mfccs, file
                lst.append(arr)
                # If the file is not valid, skip it
            except ValueError as err:
                print(err)
                continue

    print("--- Data loaded. Loading time: %s seconds ---" % (time.time() - start_time))

    # Creating X and y: zip makes a list of all the first elements, and a list of all the second elements.
    X, y = zip(*lst)
    print(lst)
    print(y)
    # Array conversion
    X, y = np.asarray(X), np.asarray(y)

    # Array shape check
    print(X.shape, y.shape)
    

    # Preparing features dump
    X_name, y_name = 'data.joblib', 'labels.joblib'

    joblib.dump(X, os.path.join(save_dir, X_name))
    joblib.dump(y, os.path.join(save_dir, y_name))

    return "Completed"
```

```python
features=r'C:\Users\ASUS\Desktop\Projects\Audio Sentiment Analysis\features'
save=r'C:\Users\ASUS\Desktop\Projects\Audio Sentiment Analysis\joblib_features'
```

```python
FEATURES = features_creator(path=features, save_dir=save)
```
